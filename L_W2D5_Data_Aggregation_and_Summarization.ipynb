```python
import numpy as np
import pandas as pd

vehicles = pd.read_csv('vehicles.csv')
vehicles.groupby(['Transmission'])
```

```python
#Here we aggregate 3 different columns and compute their mean based on the different transmission values
vehicles.groupby(['Transmission'])['Highway MPG', 'City MPG', 'Combined MPG'].mean()

vehicles.groupby(['Transmission'])['Highway MPG', 'City MPG', 'Combined MPG'].mean().reset_index()

```

```python
#In this example we aggregate based on two columns and 
#compute the median CO2 Emission for all combinations of fuel type and cylinders
#vehicles.groupby(['Fuel Type', 'Cylinders'])['CO2 Emission Grams/Mile'].median()

vehicles.groupby(['Fuel Type', 'Cylinders'])['CO2 Emission Grams/Mile'].agg(['median','mean'])

```

```python
#Here we produce the mean, median and standard deviation for combined MPG grouped by fuel type
vehicles.groupby(['Fuel Type'])['Combined MPG'].agg(['mean', 'median', 'std'])
```

```python
# custom aggregation function
from scipy import stats

def agg_mode(x):
    return(stats.mode(x)[0])
def agg_freq(x):
    return(stats.mode(x)[1])

```

```python
# show most common Vehicle Class per Transmission
vehicles.groupby("Transmission")["Vehicle Class"].agg([agg_mode, agg_freq])
#vehicles.groupby("Transmission")["Vehicle Class"].agg([stats.mode])


vehicles.groupby("Transmission")["Vehicle Class"].agg([agg_mode, agg_freq, 'min'])

```
